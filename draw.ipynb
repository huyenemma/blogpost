```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.complete_graph(15)
pos = nx.circular_layout(G)
nx.draw(G, pos, node_color='#004aad', edge_color='#ff3131', with_labels=False)

plt.figure(figsize=(8, 8))
plt.show()
```

```python
# Define the size of the lattice
rows, cols = 6, 6  # You can change this to create larger or smaller lattices

# Create a 2D grid graph
G = nx.grid_2d_graph(rows, cols)

pos = dict((n, n) for n in G.nodes())

plt.figure(figsize=(8, 8))

nx.draw(G, pos, node_color='#004aad', edge_color='#ff3131', with_labels=False, node_size=300)

plt.show()
```
